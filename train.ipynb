```python
from ultralytics import YOLO
```

```python
import shutil

shutil.copytree('/content/drive/MyDrive/all', '/content/datasets')
```

```python
model = YOLO('yolov8x.pt')

results = model.train(data='/content/drive/MyDrive/all/dataset.yaml', epochs=50)
```

```python
results = model([f'/content/drive/MyDrive/train_dataset_dataset/video0/frames_rgb/{i:>04}.png' for i in range(100)])
```

```python
from PIL import Image

for i, result in enumerate(results):
    im_array = result.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    im.save(f'result_{i:>04}.png')  # save image
```

```python
import numpy as np

images = []

for i in range(100):
    images.append(np.array(Image.open(f'result_{i:>04}.png')))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import rc

rc('animation', html='html5')

figure, axes = plt.subplots()

image_axes = []

for i, image in enumerate(images):
    if i == 0:
        im = axes.imshow(image)
    else:
        im = axes.imshow(image, animated=True)
    image_axes.append([im])

anim = animation.ArtistAnimation(figure, image_axes, interval=100, blit=True)

plt.close()
anim
```
